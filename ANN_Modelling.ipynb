# Importing necessary libraries

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('~/Downloads/Churn_Modelling.csv')
```

# Removing useless columns

```python
X=df.iloc[:,3:13]
X
```

```python
y=df['Exited']
y.head()
```

# Data Wrangling
## Encoding - LabelEncoder (Alphabetical order)
- Converting the discrete categoring data into discrete numerical values

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X["Geography"]=le.fit_transform(X['Geography'])
X
```

```python
X['Geography'].value_counts()
#France=0,Germany-1,Spain-2
```

```python
X['Gender']=le.fit_transform(X['Gender'])
X
```

```python
X['Gender'].value_counts()
#0-Female,1=Male
```

# Splitting data into test and train
- Train=75%
- Test=25%

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=True)
```

```python
X_train.shape
```

```python
X_test.shape
```

# Feature Scaling 
## StandardScaler 
- Converting the data of large as well as small scale to a same scale

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
X_train[0]
```

```python
len(X_train[0])
```

# ANN Model

# Importing keras libraries and packages

```python
pip install keras
```

```python
import keras
#For building Neural Network layer by layer
from keras.models import Sequential
```

```python
#To randomly initialize the weights to small numbers close to 0(But not 0)
from keras.layers import Dense
```

# Initializing the ANN
- 2 ways
  - Defining each layer one by one
  - Defining a graph

```python
classifier = Sequential()
```

# Adding the input layer and the first hidden layer

```python
#Activation function is Rectifier Activation Function (ReLu)
```

```python
?classifier.add
```

```python
#input layer
classifier.add(Dense(5,kernel_initializer='glorot_uniform',activation='relu',input_dim=10))
# glorot_uniform - produces random weight
# input_dim=10 as X_train and X_test contains 10 features
# 5 : Pass 5 inputs to the next layer (hidden layer)
```

```python
#Adding to hidden layer 1 (Second layer)
classifier.add(Dense(5,kernel_initializer='uniform',activation='relu'))
#Adding the output layer
classifier.add(Dense(1,kernel_initializer='uniform',activation='sigmoid'))
#Output layer has only 1 feature
#Activation is sigmoid whenever we have to use classification problem 
```

# Compile the model


```python
#Adjusting the weights done by optimizer
#loss is taken binary as it is a classification problem 
#Metrics is accuary
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

# Training model with data

```python
#batches -8000 in 10 batches so each batch as 800
#epochs means how many time we want to adjust the weight
#With every iteration we can see that loss is decreasing while accuracy is increasing
classifier.fit(X_train,y_train,batch_size=10,epochs=100)
```

# Making predictions and evaluating the model

```python
y_pred=classifier.predict(X_test)
print(y_pred)
```

```python
#Produces the probabilities
# If p>0.5 =1
# else p=0
```

```python
churn = []
churn_val =[]
for i in y_pred:
    if i > 0.5:
        churn.append("YES")
        churn_val.append(1)
    else:
        churn.append("NO")
        churn_val.append(0)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
print(confusion_matrix(y_test,churn_val))
```

```python
print("Accuracy in percentage:",accuracy_score(y_test,churn_val)*100)
```
